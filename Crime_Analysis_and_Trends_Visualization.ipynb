#Task Two

```python
# Counting the number of crimes per year
crimes_by_year = crimes_df['Year'].value_counts()

# Finding the year with the maximum number of crimes
max_crimes_year = crimes_by_year.idxmax()
max_crimes_count = crimes_by_year.max()

print(f"The year with the largest amount of crimes is {max_crimes_year} with {max_crimes_count} crimes.")
```

#Task Three

```python
# Filter data for the year 2020
crimes_2020 = crimes_df[crimes_df['Year'] == 2020]

# Count occurrences of each crime in 2020
crime_counts = crimes_2020['Primary Type'].value_counts()

# Calculate arrest rates for each crime
arrest_rates = crimes_2020.groupby('Primary Type')['Arrest'].mean()

# Find the five most common crimes in 2020
top_5_crimes = crime_counts.head(5)

# Find the crime with the highest and lowest arrest rate among the top 5 crimes
highest_arrest_rate_crime = arrest_rates[top_5_crimes.index].idxmax()
lowest_arrest_rate_crime = arrest_rates[top_5_crimes.index].idxmin()

print("The five most common crimes in 2020 are:")
print(top_5_crimes)

print("\nAmong the top 5 crimes:")
print(f"The crime with the highest arrest rate is: {highest_arrest_rate_crime} - Arrest rate: {arrest_rates[highest_arrest_rate_crime]}")
print(f"The crime with the lowest arrest rate is: {lowest_arrest_rate_crime} - Arrest rate: {arrest_rates[lowest_arrest_rate_crime]}")
```

#Task Five

```python
# Grouping by year and count the number of arrests
arrests_by_year = crimes_df.groupby('Year')['Arrest'].sum()

# Finding the year with the most number of arrests
most_arrests_year = arrests_by_year.idxmax()
num_arrests = arrests_by_year.max()

print("Year with the most number of arrests:", most_arrests_year)
print("Number of arrests made during that year:", num_arrests)

# Plotting total number of arrests per year
arrests_by_year.plot(kind='bar', figsize=(10, 6))
plt.title('Total Number of Arrests per Year')
plt.xlabel('Year')
plt.ylabel('Number of Arrests')
plt.grid(axis='y')
plt.show()
```
